```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
def merge_read_plot(file_list, file_path_and_name):
    # Step 1: Find all files matching the pattern
    file_list = glob.glob(file_list)  

    # Step 2: Read each file into a DataFrame (assuming space-separated columns)
    df_list = [pd.read_csv(file, sep='\s+', header=None) for file in file_list]  
    merged_df = pd.concat(df_list, ignore_index=True)  # Combine all DataFrames

    # Step 3: Sort DataFrame by phase (column 0)
    merged_df = merged_df.sort_values(by=[0]).reset_index(drop=True)

    # Step 4: Save sorted DataFrame
    merged_df.to_csv(file_path_and_name, index=False, sep="\t", header=False)  

    # Step 5: Read back the merged file
    df = pd.read_csv(file_path_and_name, sep='\s+', header=None)  

    # Step 6: Debugging Step - Print DataFrame Info
    print(df.head())  # Show first few rows
    print(f"DataFrame shape: {df.shape}")  # Check number of columns

    # Step 7: Ensure it has exactly 6 columns
    if df.shape[1] < 6:
        raise ValueError(f"Expected 6+ columns, but got {df.shape[1]}")

    # Step 8: Extract Columns as NumPy Arrays
    phase = df[0].to_numpy()  
    te = df[1].to_numpy() 
    flux = df[2].to_numpy()
    flux_err = df[3].to_numpy()
    weights_1 = df[4].to_numpy()
    weights_2 = df[5].to_numpy()

    # Step 9: Plot Data
    fig, ax = plt.subplots()  # Create a figure and axis
    ax.scatter(phase, flux, s=5)  # Plot phase vs flux
    ax.set_xlabel('Phase')
    ax.set_ylabel('Flux (mJsky)')
    ax.set_xlim(-0.025, 0.025)

    plt.show()

    fig2, ax2 = plt.subplots()  # Create a figure and axis
    ax2.scatter(phase, flux, s=5)  # Plot phase vs flux
    ax2.set_xlabel('Phase')
    ax2.set_ylabel('Flux (mJsky)')
    ax2.set_xlim(0.48, 0.52)

    plt.show()

    return phase, te, flux, flux_err, weights_1, weights_2
```

```python
file_list = "/home/pha19jgb/LCURVE/U/*.txt"
file_path_and_name = "/home/pha19jgb/LCURVE/U/U.txt"

phase, te, flux, flux_err, weights_1, weights_2 = merge_read_plot(file_list, file_path_and_name)
```
